```python
# Input marks for three tests
m1 = int(input("Enter marks test 1: "))
m2 = int(input("Enter marks test 2: "))
m3 = int(input("Enter marks test 3: "))

# Calculate the average of the best two marks
if m1 <= m2 and m1 <= m3:
    avg = (m2 + m3) // 2
elif m2 <= m1 and m2 <= m3:
    avg = (m1 + m3) // 2
else:
    avg = (m1 + m2) // 2

# Output the average
print("Average of best 2 test marks is: ", avg)
```

```python
def palindrome(num):
    str_num = str(num)
    return str_num == str_num[::-1]

num = int(input("Enter a number"))
print("Palindrome" if palindrome(num) else "Not a palindrome")
str_num = str(num)
for i in range(0, 10):
    if str_num.count(str(i)) > 0:
        print(str(i), " -> ", str_num.count(str(i)))
```

```python
def func(num):
    if(num == 1):
        return 0
    if(num == 2):
        return 1
    return func(num - 1) + func(num - 2)

num = int(input("Enter a number: "))
if num > 0:
    print("Function calls value: ", func(num))
else:
    print("Invalid input")
```

```python
def binary_to_decimal(binary):
    decimal = int(binary, 2)
    return decimal

def octal_to_hexdecimal(octal):
    decimal = int(octal, 8)
    hexadecimal = hex(decimal)[2:]
    return hexadecimal.upper()

binary = input("Enter a binary number")
print("Binary number to Decimal number: ", binary_to_decimal(binary))

octal = input("Enter a octal number")
print("Octal number to Hexadecimal number: ", octal_to_hexdecimal(octal))
```

```python
sentence = input("Enter a sentence")
word_count = 0
digit_count = 0
upper_count = 0
lower_count = 0
word_count = len(sentence.split())

for ch in sentence:
    if ch.isdigit():
        digit_count += 1
    if ch.islower():
        lower_count += 1
    if ch.isupper():
        upper_count += 1

print("Nummber of words: ",word_count)
print("Nummber of digits: ",digit_count)
print("Nummber of lower case letters: ",lower_count)
print("Nummber of upper case letters: ",upper_count)
```

```python
import difflib
str1 = input("Enter the first string")
str2 = input("Enter the second string")

seq = difflib.SequenceMatcher(None, str1, str2)
diff = round(seq.ratio() * 100, 2)

print(str(diff), "% Match between the strings")
```

```python
from bokeh.plotting import figure, output_file, show 
x= [val for val in range(10)]
y = [val for val in range(0, 20, 2)]
output_file("basiclegend.html")
p = figure()
p.line(x, y, legend_label="My Red Line", line_color="red") 
p.line(y, x, legend_label="My Orange Line", line_color="orange")
p.line(y[::-1], x, legend_label="My Green Line", line_color="green")
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
p = figure()

x= [x for x in range(1, 11)]
colors = ['red', 'green', 'blue', 'yellow']
for i in range(2, 6):
    p.line(x, [val*i for val in x], line_width=2, color=colors[i-2], alpha=0.8, legend_label='Multiples of {}'.format(i))
    
p.legend.location = "top_left"
p.legend.click_policy = "hide"
output_file("interactive _legend.html")
show(p)
```

```python
from bokeh.io import output_file, show
from bokeh.layouts import row
from bokeh.plotting import figure

# Set the output file
output_file("line_circle_plots.html")

# Create the first figure
fig1 = figure(width=400, height=400, title="Plot 1")
fig1.line([2, 1, 5, 3, 4, 7, 6], [1, 4, 3, 5, 2, 7, 7], line_width=4)

# Create the second figure
x = y = list(range(10))
fig2 = figure(width=400, height=400, title="Plot 2")
fig2.circle(x, y, size=5)

# Show both figures in a row layout
show(row(fig1, fig2))
```

```python
import numpy as np
import plotly.graph_objects as go

# Create x, y, and z values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)  # Create a grid
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))  #Compute z values

# Create the surface plot
surface_plot = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
surface_plot.show()
```
